# Project 1

In this project, Motor Vehicle Collisions - Crashes data from NYC open data.

The goal is to study the number of person injured from collosions in NYC in 2024. 

The original data has more than 2M rows, however, I filtered only 2024 to get 20549 rows. 


```python
import pandas as pd

# Load the dataset
df = pd.read_csv("MV-Coll_Crash_2024.csv")

# Display the first few rows
print(df.head())

# Check column names and data types
print(df.info())
```

```python
col = "NUMBER OF PERSONS INJURED"
```

```python
mean_val = df[col].mean()
median_val = df[col].median()
mode_val = df[col].mode()[0]

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```

```python
injury_rate = (df[col] > 0).mean() * 100
print(f"{injury_rate:.1f}% of crashes involved at least one injury.")
```

Now I will use only python library 

```python
import csv
from collections import Counter

injuries = []

with open("MV-Coll_Crash_2024.csv", newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        val = row["NUMBER OF PERSONS INJURED"]
        if val.isdigit():
            injuries.append(int(val))

# Mean
mean_val = sum(injuries) / len(injuries)

# Median
injuries_sorted = sorted(injuries)
n = len(injuries_sorted)
if n % 2 == 1:
    median_val = injuries_sorted[n // 2]
else:
    median_val = (injuries_sorted[n // 2 - 1] + injuries_sorted[n // 2]) / 2

# Mode
mode_val = Counter(injuries_sorted).most_common(1)[0][0]

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```
